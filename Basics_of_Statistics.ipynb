1. Explain the different types of data(qualitative and quantitative) and provide examples of each. Discuss nominal, ordinal, interval, and ratio scales.

 Qualitative Data - It is categorical. We can not do mathematical operations in this type of data. Each data focuses on descriptive information rather than numbers. For example - Customer feedback, User Experience (UX) Design, blood group, Gender etc.

 Quantitative data - This type of data can be measured or calculated. It is in numerical form. For example - Customer survey: 4 out of 5 customers rated the service as excellent, Education: The student scored 92% on the final exam etc.

 Different types of scales of measurement are mentioned below:

 1. Nominal(Qualitative or categorical data)- No order in the data Eg: Gender, Ethnicity.
 2. Ordinal Scale- Order & rank has a meaning/matter. Difference can't be measured. We can make pie chart/ Bar plot through this type of data. Eg: Performance of students(1st rank, 2nd rank or 3rd rank).
 3. Interval Scale data- The rank and order has a meaning. Difference can be measured (excluding ratio). It doesn't have compulsorily 0 starting value. Eg: Length, Weight, Height, Temp etc.
 4. Ratio scale- Order & rank have a meaning. Difference and ratio are measurable. It has a 0 starting point compulsorily.

```python
#Mode >> It counts maximum frequency(repeated highest no. of times).
import scipy as sc
d=[2,4,8,20,10,100,1,2,3,3,6,4,5,4,51,2,3,3,6,4,5,4,100]
sc.stats.mode(d)
```

```python
#Use cases of central tendency
import pandas as pd
df=pd.DataFrame({'Age':[25,26,None,23,25], 'Gender':['M','F','M',None,'M'], 'Weight': [80,70,30,None,1000], 'Salary(k)': [None, 50, 60, 70, 45]})
df
```

```python
#Age is continuous variable so find 2nd no. age by mean
df['Age'].fillna(df['Age'].mean(), inplace=True)
df
#Gender is categorical data so find 3rd no. gender by mode
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df
#In Weight column, outlier is available so find it by median
df['Weight'].fillna(df['Weight'].median(), inplace=True)
df
#Salary is continuous variable so find 4th no. salary by mean
df['Salary(k)'].fillna(df['Salary(k)'].mean(), inplace=True)
df
```

3. Explain the concept of dispersion. How do variance and standard deviation measure the spread of data?

 Dispersion refers to the way data values are spread out or scattered around a central value (like the mean or median). It tells us how much the data varies—whether the values are tightly clustered or widely spaced. Two common statistical tools used to measure dispersion are:

 1. Variance: This measures the average squared deviation of each data point from the mean.
 2. Standard Deviation: It is the square root of the variance.

 Eg: Set A: 85, 86, 84, 85, 85 → Low dispersion
     
     Set B: 70, 90, 60, 95, 85 → High dispersion

     Both might have the same mean, but Set B will have a higher variance and standard deviation because the scores are more spread out.

4. What is a box plot, and what can it tell you about the distribution of data?

 A box plot is also called a box-and-whisker plot and it is a visual summary of a dataset’s distribution, showing its central tendency, spread, and potential outliers.

 What it tells you: It tells you the median where the middle of the data lies, the range and IQR, how spread out the data is, if the median is closer to the top or bottom of the box, or if one whisker is longer, the data might be skewed and points beyond the whiskers highlight unusual values.

5. Discuss the role of random sampling in making inferences about populations.

 Random sampling plays a critical role in making accurate and unbiased inferences about a population. It ensures that every individual in the population has an equal chance of being selected. This helps the sample reflect the diversity and structure of the whole population, making conclusions more generalizable. By avoiding systematic selection, random sampling reduces selection bias, which can skew results and lead to false conclusions. Statistical methods rely on random samples to estimate population parameters (like mean or proportion) and to calculate margins of error or p-values.

6. Explain the concept of skewness and its types. How does skewness affect the interpretation of data?

 Skewness is a statistical measure that describes the degree of asymmetry of a distribution around its mean. In simple terms, whether the data is spread out more to the left or to the right of the average value. Types of Skewness: 1. Right-Skewed- It is also called as Positive skewed, tail on the right side is longer or fatter. Mean > Median > Mode. 2. Left-Skewed: It is also called as Negative skewed, tail on the left side is longer or fatter. Mean < Median < Mode. 3. Symmetrical Distribution: It is also called as Zero Symmetric, that distribution is perfectly symmetrical. Mean = Median = Mode.

7. What is the interquartile range(IQR), and how is it used to detect outliers?

 The interquartile range (IQR) is a measure of statistical dispersion, or how spread out the values in a data set are. It specifically focuses on the middle 50% of the data. ( IQR= Q3-Q1 )

 Detecting Outliers with IQR:
 1. Calculate Q1 and Q3
 2. Compute IQR = Q3 - Q1
 3. Determine the lower bound: Q1 - 1.5 × IQR
 4. Determine the upper bound: Q3 + 1.5 × IQR
 5. Any data point below the lower bound or above the upper bound is considered an outlier.

11. Explain what a random variable is and differentiate between discrete and continuous random variables.

 A random variable is a numerical quantity whose value is determined by the outcome of a random phenomenon. In simpler terms, it's a variable that takes on different values randomly.

 Discrete random variable- It takes on a countable number of distinct values. Eg- The number of heads in 3 coin tosses (values: 0, 1, 2, 3).

 Continuous random variable- It takes on uncountably infinite values over a range or interval. Eg- The height of students in a class, The time it takes to run a mile.
